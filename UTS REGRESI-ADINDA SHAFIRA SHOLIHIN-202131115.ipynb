# PREPOCESSING DATA

```python
from sklearn.preprocessing import LabelEncoder
for col in ['Study Hours (X)','Exam Scores (Y)']:
    le= LabelEncoder()
    data[col]= le.fit_transform(data[col])
    data
    print (data['Study Hours (X)'],data['Exam Scores (Y)'])
    # LabelEncoder dari pustaka scikit-learn untuk mengonversi nilai kategori dalam dua kolom, yaitu 'Study Hours (X)' dan 'Exam Scores (Y)', dari bentuk kategorikal menjadi representasi numerik.
```

```python
plt.scatter(data['Study Hours (X)'], data['Exam Scores (Y)'])
#Membuat scatter plot dengan menggunakan data pada kolom 'Study Hours (X)' sebagai sumbu X dan 'Exam Scores (Y)' sebagai sumbu Y.
plt.title('Study Hours vs Exam Scores')
#Memberikan judul pada plot, dalam hal ini, "Study Hours vs Exam Scores".
plt.xlabel('Lama Belajar (Jam)')
#Menyertakan label pada sumbu X, yaitu "Lama Belajar (Jam)".
plt.ylabel('Nilai Ujian') 
#Menyertakan label pada sumbu Y, yaitu "Nilai Ujian".
plt.show()
#Menampilkan plot.
```

```python
from sklearn.model_selection import train_test_split

X = data['Study Hours (X)'].values.reshape(-1,1)
Y = data['Exam Scores (Y)'].values

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=.2)
#Membagi data menjadi set pelatihan dan pengujian menggunakan train_test_split.
```

# REGRESI

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)
#Membuat model regresi linier dan melatihnya menggunakan data pelatihan.
```

# SPLIT KOLOM

```python
#membagi data x dan y
study_hrs = data['Study Hours (X)'].values[:, np.newaxis]#sebagai x_train
exam_scr = data['Exam Scores (Y)'].values#sebagai y_train
#tidak pakai split train karena datanya sedikit
```

```python
study_hrs#menampilkan data 
```

# MODELLING

```python
from sklearn.linear_model import LinearRegression
#Mengimpor kelas LinearRegression dari modul sklearn.linear_model. Kelas ini menyediakan implementasi model regresi linear.
model1 = LinearRegression()
#Membuat objek model regresi linear menggunakan kelas LinearRegression(). Objek ini nantinya akan digunakan untuk melatih dan membuat prediksi dengan model regresi linear.
model1.fit(study_hrs, exam_scr)
#Melatih model regresi linear dengan data latih. Parameter study_hrs dijadikan sebagai variabel independen (X), dan exam_scr sebagai variabel dependen (Y). Dengan memanggil metode fit(), model akan belajar dari data latih untuk menyesuaikan garis regresi linear yang dapat memodelkan hubungan antara jumlah jam belajar dan nilai ujian.
```

```python
predik = model.predict(x_test)
predik
#membuat prediksi dari x_test
```

# TESTING

```python
study_hrs_test =[[3],[9]]
pred_exam_scr = model1.predict(study_hrs_test)
print("Prediksi Exam Score: ", pred_exam_scr)
#mencoba memperoleh prediksi nilai ujian untuk dua skenario jumlah jam belajar: 3 jam dan 9 jam. Hasil prediksi ini dapat memberikan perkiraan nilai ujian berdasarkan model regresi linear yang telah dilatih sebelumnya.
```

# VISUALISASI

```python
plt.scatter(x_test, y_test, c='blue')
plt.plot(x_test, predik, c='red')
plt.show()
#menampilkan scatter plot data pengujian (biru) dan garis regresi hasil prediksi (merah).
```
